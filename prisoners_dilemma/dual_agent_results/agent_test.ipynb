```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from utils import sweep_test, construct_2
from matplotlib.colors import LinearSegmentedColormap
cmap0 = LinearSegmentedColormap.from_list("", ["white", "black"])
plt.rcParams['figure.facecolor'] = 'white'
plt.rcParams['axes.facecolor'] = 'white'
plt.rcParams['savefig.facecolor'] = 'white'
```

```python
T = 300
agent_1, agent_2, D = construct_2(lr_pB = 0.2,lr_pB_2 = 0.2, factors_to_learn="all")
from pymdp import utils
C = utils.obj_array(1)
C[0] = np.array([3, 1, 4, 2])
agent_1.C = C
agent_2.C = C

actions_over_time, B1_over_time, q_pi_over_time, q_s_over_time = sweep_test(agent_1, agent_2, observation_1 = [0], observation_2 = [0],D=D,T=T, test_start_time=20,test_end_time=200,test_interval_time=4)
```

```python

fig, ax = plt.subplots(figsize = (16,15))
ax.grid(color='w', which = 'minor', linestyle='-', linewidth=2)
ax.set_xlabel("Time")
im = ax.imshow(actions_over_time.T,cmap ='gray', aspect = 10, vmin=0, vmax =1)
ax.set_xlabel("Time",fontsize = 12)
ax.set_xticks(np.linspace(0,T,25).round(0),fontsize = 12)
ax.set_yticks([0, 1], labels=["Agent 1", "Agent 2"],fontsize = 12)
plt.savefig('testing')
plt.show()
```

```python
def set_labels(axis):
    axis.set_xticklabels(['CC','CD','DC','DD'],fontsize =14)
    axis.set_yticklabels(['CC','CD'],fontsize =14)
    axis.set_xlabel(r'$s_t$',fontsize =20)

    axis.set_ylabel(r'$s_{t+1}$',fontsize =20)

def set_labels_2(axis):
    axis.set_xticklabels(['CC','CD','DC','DD'],fontsize =14)
    axis.set_yticklabels(['DC','DD'],fontsize =14)
    axis.set_xlabel(r'$s_t$',fontsize =20)

    axis.set_ylabel(r'$s_{t+1}$',fontsize =20)

fig, ax = plt.subplots(3,2, figsize = (10,9))

img1 = sns.heatmap(B1_over_time[10,:2,:,0,0],ax =ax[0,0], cmap = cmap0, vmin = 0, vmax =1)
img2 = sns.heatmap(B1_over_time[10,2:,:,1,0],ax =ax[0,1], cmap = cmap0, vmin = 0, vmax =1)
img3 = sns.heatmap(B1_over_time[100,:2,:,0,0],ax =ax[1,0], cmap = cmap0, vmin = 0, vmax =1)
img4 = sns.heatmap(B1_over_time[100,2:,:,1,0],ax =ax[1,1], cmap = cmap0, vmin = 0, vmax =1)
img5 = sns.heatmap(B1_over_time[299,:2,:,0,0],ax =ax[2,0], cmap = cmap0, vmin = 0, vmax =1)
img6 = sns.heatmap(B1_over_time[299,2:,:,1,0],ax =ax[2,1], cmap = cmap0, vmin = 0, vmax =1)

ax[0,0].set_title('t= 10, u = C',fontsize =20)
ax[0,1].set_title('t= 10, u = D',fontsize =20)

ax[1,0].set_title('t= 100, u = C',fontsize =20)
ax[1,1].set_title('t= 100, u = D',fontsize =20)
ax[2,0].set_title('t= 500, u = C',fontsize =20)
ax[2,1].set_title('t= 500, u = D',fontsize =20)

set_labels(ax[0,0])
set_labels_2(ax[0,1])
set_labels(ax[1,0])
set_labels_2(ax[1,1])
set_labels(ax[2,0])
set_labels_2(ax[2,1])

fig.tight_layout()
plt.savefig("Learned Transitions_interval10")

```
